```python
import numpy as np 
import matplotlib.pyplot as plt
import os
```

```python
dataset0 = np.load("E:\PhD_Works\Courses\Security_course_2022\SNR_\dataset\Alignment\Alignment0.npy")
print(dataset0.shape)
```

```python
plt.figure(figsize=(14,5))
print(dataset0.shape)
for i in range (0, dataset0.shape[0]):
    plt.plot(dataset0[i][1200:1400], label='trace')


plt.title('Trace Alignment checking')
plt.xlabel('sample')
plt.ylabel('Power consumption')
plt.legend()
plt.savefig("Alignment0_zoomed.pdf")
plt.show()
```
